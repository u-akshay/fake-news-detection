The Dataset have 4 columns.
    + Unnamed: 0
    + title
    + text
    + label
`Unnamed: 0` is  is seems unwanted datas, hence it removed for further work.

`title` is the News Title. I think it has no role in the prediction, so i just avoided it.

`text` and `label` are the main features used for the prediction.

```python
#make X
X = df['text']
X
```

```python
#make y
y = df['label']
y
```

```python
#count vectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#var of CountVect
vect = CountVectorizer(stop_words='english')
```

```python
#fit function of vect and saved it into 'vectorizer' for make joblib
vectorizer = vect.fit(X)
```

```python
vect.vocabulary_
```

```python
#spliting library
from sklearn.model_selection import train_test_split
```

```python
#split for prediction
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=99)
```

```python

X_train_transformed = vect.transform(X_train)
X_test_transformed = vect.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(X_train_transformed,y_train)
```

```python
X_train_transformed
```

```python
y_train
```

```python
prediction = mnb.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(prediction,y_test)
```

### now train with full set of data -- 100% of data and store it for future prediction

```python
X_train_transformed = vect.transform(X)
```

```python
naive = mnb.fit(X_train_transformed,y)
```

```python
import joblib
```

```python
joblib.dump(naive,"naive.joblib")
joblib.dump(vectorizer,"vectoriszer.joblib")
```

```python
#accuracy score of first 100 random state with test size 20%

for i in range (100):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=i)
    X_train_transformed = vect.transform(X_train)
    X_test_transformed = vect.transform(X_test)
    mnb = MultinomialNB()
    naive = mnb.fit(X_train_transformed,y_train)
    prediction = mnb.predict(X_test_transformed)
    print(i,"====",accuracy_score(prediction,y_test))
```
